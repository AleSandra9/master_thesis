```python
'''Import ticker list'''

etf_list = pd.read_csv("/Users/alexandra/Documents/master_thesis/data/sb_etfs.csv")

etf_list
```

```python
"""
Get adjusted stock prices for SB ETFs.
"""

tickers = list(etf_list['Ticker'])
etfs=[]
for ticker in tickers:
    etf = YahooDailyReader(ticker, start=datetime.date(2002, 12, 1),interval='d').read()
    etfs.append(etf)
```

```python
etfs
```

```python
etf_prices=pd.concat([etf.iloc[:,-1] for etf in etfs],axis=1)
```

```python
etf_prices.columns=tickers
etf_prices
```

```python
etf_prices.to_csv("/Users/alexandra/Documents/master_thesis/etfs_prices_d.csv")

'''Resampling monthly prices'''
etf_prices = etf_prices.resample('BM').last()
etf_prices=etf_prices[0:-3]
etf_prices
```

```python
'''Save adjusted close prices'''
etf_prices.to_csv("/Users/alexandra/Documents/master_thesis/etfs_prices.csv")
```
